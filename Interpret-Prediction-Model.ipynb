## Get Datasets

```python
input_dir = 'Predict-Datasets'
output_dir = 'Resized-Predict-Datasets'
target_size = (28, 28)

# Create the output directory if it doesn't exist
os.makedirs(output_dir, exist_ok=True)

image_number = 1
# Iterate over the files in the directory
while os.path.isfile(f"{input_dir}/Predict-{image_number}.jpg"):
    # Load the image
    img_path = f"{input_dir}/Predict-{image_number}.jpg"
    img = Image.open(img_path)
        
    # Resize the image while maintaining the aspect ratio using thumbnail method
    img.thumbnail(target_size, Image.ANTIALIAS)
        
    # Create a new image with the target size as canvas
    resized_image = Image.new('L', target_size, 255)
        
    # Paste the resized image onto the canvas
    offset = ((target_size[0] - img.size[0]) // 2, (target_size[1] - img.size[1]) // 2)
    resized_image.paste(img, offset)
        
    # Save the resized image to the output directory
    output_path = f"{output_dir}/Predict-{image_number}.jpg"
    resized_image.save(output_path)
        
    image_number += 1
```

## Resize Image

```python
import cv2

# Set the path to our handwriting dataset
dataset_dir = 'Resized-Predict-Datasets'

images = []

# Each image file in the writer directory corresponds to a handwriting sample
for image_file in os.listdir(dataset_dir):
    image_path = os.path.join(dataset_dir, image_file)
    img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
    
    # Normalize the image by dividing by the maximum pixel value (255)
    img_normalized = img.astype('float32') / 255.0
    
    # Append the normalized image to the list
    images.append(img_normalized)

# Load the handwriting data 
# Convert the lists to numpy arrays
images = np.array(images)
```

## tf.explain

```python
import cv2
import tensorflow as tf
import matplotlib.pyplot as plt
from tf_explain.core.grad_cam import GradCAM

# Set the path to your saved model
model_path = 'writerIdentifier_model.h5'

# Load the model
model = tf.keras.models.load_model(model_path)

# Set the path to your images directory
images_dir = 'Resized-Predict-Datasets'

# Define the grid size
grid_rows = 2
grid_cols = 3

# Calculate the total number of images
total_images = len(os.listdir(images_dir))

# Adjust the grid size if necessary
if total_images > grid_rows * grid_cols:
    # Calculate the required number of rows and columns
    grid_rows = (total_images - 1) // grid_cols + 1

# Create a figure and set the size
fig = plt.figure(figsize=(16, 12))

# Iterate over the images in the directory
for i, image_file in enumerate(os.listdir(images_dir)):
    # Read the image
    image_path = os.path.join(images_dir, image_file)
    img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
    
    # Preprocess the image
    img = img.astype('float32') / 255.0
    img = np.expand_dims(img, axis=0)
    
    # Generate predictions
    predictions = model.predict(img)
    predicted_class = np.argmax(predictions)
    
    # Create a GradCAM explainer
    explainer = GradCAM()

    # Compute the heatmap
    heatmap = explainer.explain((img, None), model, class_index=predicted_class)
    
    # Create a subplot at the specified grid position
    ax = plt.subplot2grid((grid_rows, grid_cols), (i // grid_cols, i % grid_cols))
    
    # Map the predicted index to the actual writer label
    # 1 - Janice; 2 - Jasmine
    writers = [1, 2]  # List of writer labels used during training
    predicted_writer = writers[predicted_class]
    
    # Plot the heatmap
    ax.imshow(heatmap, cmap='jet')
    ax.axis('off')
    ax.set_title(f'Heatmap - Class {"Janice" if predicted_writer == 1 else "Jasmine"}')

# Adjust the layout and display the figure
plt.tight_layout()
plt.show()
```
